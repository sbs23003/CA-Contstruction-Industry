```python
def plot_country_data(df, country):
    # Filter the data for the specified country
    country_data = df[df['COUNTRY'] == country]

    # Exclude the 'COUNTRY', 'Date', 'Number_of_Companies', and 'Number_of_Employees' columns from plotting
    columns_to_plot = country_data.columns[2:].difference(['Number_of_Companies', 'Number_of_Emplyees'])

    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 8))

    # Plot each column
    for column in columns_to_plot:
        ax.plot(country_data['Date'], country_data[column], label=column)

    # Set the title and labels
    ax.set_title(country)
    ax.set_xlabel('Date')
    ax.set_ylabel('Value')

    # Add a legend
    ax.legend()

    # Show the plot
    plt.show()
```
